# Data  Transform and cleaning

```python
# Load the dataset
df = pd.read_csv(r"./DataSet/IPL_Ball_by_Ball_2008_2022.csv")

# Remove some features (columns) from the dataframe
df = df.drop(['ballnumber', 'batter', 'bowler', 'non-striker', 'extra_type', 'batsman_run', 'extras_run',
              'non_boundary', 'player_out', 'kind', 'fielders_involved'], axis=1)

# Group the dataset by one or more columns
grouped_data = df.groupby(['ID', 'innings', 'BattingTeam'])

# Apply one or more aggregation functions to the grouped data
result_df = grouped_data.agg({'total_run': 'sum', 'isWicketDelivery': 'mean', 'overs': 'max'})

# Rename the columns of the new dataframe
result_df = result_df.rename(columns={'total_run': 'TotalRuns', 'isWicketDelivery': 'AvgWickets', 'overs': 'MaxOvers'})

# Reset the index of the new dataframe
result_df = result_df.reset_index()

# Store the new dataframe as a CSV file
result_df.to_csv('cleaned_data.csv', index=False)

result_df
```

```python
df2 = pd.read_csv(r"./DataSet/IPL_Matches_Result_2008_2022.csv")
df2 = df2.drop(['City', 'Date', 'Season', 'MatchNumber', 'SuperOver', 'WonBy', 'Margin', 'method',
               'Player_of_Match', 'Umpire1', 'Umpire2'], axis=1)
df2.info()
```

```python
# Load the dataset
df = pd.read_csv(r"./DataSet/IPL_Ball_by_Ball_2008_2022.csv")

# Remove some features (columns) from the dataframe
df = df.drop(['ballnumber', 'batter', 'bowler', 'non-striker', 'extra_type', 'batsman_run', 'extras_run',
              'non_boundary', 'player_out', 'kind', 'fielders_involved'], axis=1)


# Group the dataset by one or more columns
grouped_data = df.groupby(['ID', 'innings', 'BattingTeam'])

# Apply one or more aggregation functions to the grouped data
result_df = grouped_data.agg({'total_run': 'sum'})

# Rename the columns of the new dataframe
result_df = result_df.rename(columns={'total_run': 'TotalRuns'})

# Reset the index of the new dataframe
result_df = result_df.reset_index()

result_df = pd.merge(result_df, df2, on='ID')

# Store the new dataframe as a CSV file
result_df.to_csv('cleaned_data2.csv', index=False)

result_df
```
